```python
import pandas as pd
import numpy as np
```

```python
missing_sectors = []

missing_sectors_2 = []

# ["DDS_", "DY_", "MIGT_", "DFFD_"]

model_DDS = "DDS_"
model_DY = "DY_"
model_MIGT = "MIGT_"
model_DFFD = "DFFD_"
```

# For DDS


```python
mdcoefdf = pd.read_csv("./new_data/lumberton/Lumberton_DDS.csv")
mdcoefdf
```

```python
baseKapSectors = pd.read_csv("./new_data/lumberton/baseKap.csv")
sectorlist_set = set(baseKapSectors["sector"].unique().tolist())


modelcoeff_sectors = set(mdcoefdf.columns[5:])
model_subModel_sectors = set([s.split(" ")[-1] for s in mdcoefdf["Model_Name"]])
missing_sub_sectors = list(sectorlist_set - model_subModel_sectors)
print(missing_sub_sectors)
missing_sectors = list(sectorlist_set - modelcoeff_sectors)
print(missing_sectors)
```

```python
df_bottom_left = pd.DataFrame(
    [
        ["Galveston", "hurricane", "sector_model", f"{model_DDS} {sect}"]
        for sect in missing_sectors_2
    ],
    index=range(len(missing_sectors_2)),
    columns=mdcoefdf.columns[:4],
)
df_bottom_left
```

```python
df_right = pd.DataFrame(
    np.float64(0), index=range(mdcoefdf.shape[0]), columns=missing_sectors
)
df_right
```

```python
mdcoefdf = pd.concat([mdcoefdf, df_right], axis=1)
mdcoefdf
```

```python
df_bottom_right = pd.DataFrame(
    np.float64(0), index=range(len(missing_sectors_2)), columns=mdcoefdf.columns[4:]
)
df_bottom_right
```

```python
df_bottom = pd.concat([df_bottom_left, df_bottom_right], axis=1)
df_bottom
```

```python
final_df = pd.concat([mdcoefdf, df_bottom], axis=0).reset_index(drop=True)
```

```python
final_df
```

```python
final_df.to_csv("./new_data/Galveston/updated_coeff/DDS_coefficients.csv", index=False)
```
